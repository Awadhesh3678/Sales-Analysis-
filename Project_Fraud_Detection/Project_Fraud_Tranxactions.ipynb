```python
import pandas as pd
import matplotlib as plt

df= pd.read_csv("Raw_OnlineTransactions.csv")
df.head(10)
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    duplicate_count = df.duplicated(subset=[col]).sum()
    print(f"Column '{col}' has {duplicate_count} duplicate values.")

```

```python
df = df.drop_duplicates(subset='TransactionID')
```

```python
df.describe(include="all")
```

```python
fraud_count=df[df['IsFraud']==1].shape[0]
print(f"Total number of fraud transactions: {fraud_count}")
    
```

```python
# Fill missing Location with 'Unknown'
df['Location'] = df['Location'].fillna('Unknown')
```

```python
# Fill missing Device with 'Unknown'
df['Device'] = df['Device'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
missing_fraud_count = df[(df['IsFraud'] == 1) & (df['Amount'].isna())].shape[0]
print(f"Number of fraud transactions with missing Amount: {missing_fraud_count}")
```

```python
# Remove rows where Amount is missing or 'unknown'
df = df[df['Amount'].notnull()]
df = df[df['Amount'] != 'unknown']
```

```python
df.isnull().sum()
```

```python
df.describe(include="all")
```

```python
df.to_csv("Cleaned_OnlineTransactions.csv", index=False)
```
